```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import seaborn as sns


# 설치된 나눔폰트 경로 확인
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

# 폰트 등록
fm.fontManager.addfont(font_path)
nanum_font = fm.FontProperties(fname=font_path).get_name()

# 기본 폰트 설정
plt.rcParams['font.family'] = nanum_font
plt.rcParams['axes.unicode_minus'] = False
```

### 연도별 제품 매출 시각화 실습
이 노트북은 리스트 기반으로 생성한 가상의 연도별 A/B/C 제품 매출 데이터를 활용하여 Matplotlib과 Seaborn으로 다양한 형태의 그래프를 그려보는 실습입니다. 모든 시각화는 pandas의 DataFrame 없이 리스트만을 사용합니다. 시각화는 단일 변수부터 복수 변수 비교까지 점차 확장되며, 그래프의 색상과 선 스타일도 함께 지정합니다.


```python
# 연도별 A, B, C 제품 매출액 리스트 정의
years    = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
a_sales  = [500, 530, 560, 600, 640, 680, 720, 760, 800, 850, 900]
b_sales  = [300, 330, 390, 450, 500, 560, 600, 650, 700, 750, 800]
c_sales  = [700, 680, 660, 640, 620, 600, 590, 580, 570, 560, 550]
```

### 1. A제품 매출액 선그래프 (Matplotlib)

```python
plt.plot(years,a_sales)
plt.show()
```

```python
plt.figure(figsize=(8, 4)) #figure 크기설정
plt.plot(years, a_sales) #연도별 A제품 매출 데이터 선 그래프
plt.title('연도별 A제품 매출 추이 (Matplotlib)') #그래프 제목 설정
plt.xlabel('연도') # x축 라벨 설정
plt.ylabel('A제품 매출액') # y축 라벨 설정
plt.grid(True) # 격자선 표시
plt.show()
```

image.png

```python
plt.figure(figsize=(8, 4)) #figure 크기설정
plt.plot(years, a_sales, label="A제품") #연도별 A제품 매출 데이터 선 그래프
plt.title('연도별 A제품 매출 추이 (Matplotlib)') #그래프 제목 설정
plt.xlabel('연도') # x축 라벨 설정
plt.ylabel('A제품 매출액') # y축 라벨 설정
plt.legend()
plt.xticks([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.grid(True) # 격자선 표시
plt.show()
```

### 2. B제품 매출액 선그래프 (Seaborn)

```python
sns.lineplot(x=years, y=b_sales)
plt.show()
```

```python

plt.figure(figsize=(8, 4))
sns.lineplot(x=years, y=b_sales,label="B제품")
plt.title('연도별 B제품 매출 추이 (Seaborn)')
plt.xlabel('연도')
plt.ylabel('B제품 매출액')
plt.xticks([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.grid(True)
plt.show()
```

### 3. C제품 매출액 막대그래프 (Matplotlib)

```python
plt.bar(years,c_sales)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(years, c_sales)
plt.title('연도별 C제품 매출 (Matplotlib)')
plt.xlabel('연도')
plt.ylabel('C제품 매출액')
plt.xticks([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.grid(True, axis='y')
plt.show()
```

### 4. A제품 매출액 막대그래프 (Seaborn)

```python
sns.barplot(x=years, y=a_sales)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=years, y=a_sales)
plt.title('연도별 A제품 매출 (Seaborn)')
plt.xlabel('연도')
plt.ylabel('A제품 매출액')
plt.grid(True, axis='y')
plt.show()
```

### 5. A/B/C 제품 매출을 한 그래프에 그리기 (Matplotlib)

```python
plt.plot(years, a_sales)
plt.plot(years, b_sales)
plt.plot(years, c_sales)
plt.show()
```

각 제품의 매출 흐름을 비교할 수 있도록 색상, 선 스타일, 마커를 다르게 설정하여 시각적으로 구분할 수 있도록 추가합니다.

```python
plt.figure(figsize=(10, 5))
plt.plot(years, a_sales, marker='o', linestyle='-', color='blue', label='A제품')
plt.plot(years, b_sales, marker='s', linestyle='--', color='green', label='B제품')
plt.plot(years, c_sales, marker='^', linestyle='-.', color='orange', label='C제품')
plt.title('연도별 제품 매출 비교 (Matplotlib)')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.xticks([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.legend()
plt.grid(True)
plt.show()
```

### 6. A/B/C 제품 매출 비교 (Seaborn)

다음 조건에 맞는 선 그래프(line plot) 를 Seaborn으로 작성하세요.


1. A제품은 보라색(purple), 마름모 마커('D'), 실선(':')으로 표시할 것

2. B제품은 빨간색(red), x 마커('X'), 실선('-')으로 표시할 것

3. C제품은 회색(gray), 육각형 마커('h'), 점-선('--')으로 표시할 것

4. 제목은 '연도별 제품 매출 비교 (Seaborn)'으로 설정

5. x축에 전체 연도가 출력되도록 설정

6. x축은 '연도', y축은 '매출액'으로 라벨링

6. 범례(legend)와 격자선(grid) 포함



**아래와 같은 그래프를 출력**

image.png

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=years, y=a_sales, marker='D', linestyle=':', color='purple', label='A제품')
sns.lineplot(x=years, y=b_sales, marker='X', linestyle='-', color='red', label='B제품')
sns.lineplot(x=years, y=c_sales, marker='h', linestyle='--', color='gray', label='C제품')
plt.title('연도별 제품 매출 비교 (Matplotlib)')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.xticks([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025])
plt.grid(True)
plt.show()
```
