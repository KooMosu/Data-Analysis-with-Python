## 데이터 변환하기

### '일시' 컬럼 삭제  
데이터프레임에 '일시'라는 이름의 컬럼이 존재한다면 해당 컬럼을 삭제하세요.  
분석에 필요 없는 불필요한 열은 미리 제거하는 것이 좋습니다.

```python
df=df.drop(["일시"],axis=1)
```

## 데이터 정제하기

### 결측치 확인  
각 컬럼별 결측치(null 또는 NaN)가 몇 개인지 확인하세요.  
결측치가 많은 컬럼은 분석에 영향을 줄 수 있으므로 사전에 파악하는 것이 중요합니다.


```python
df.isnull().sum()
```

### 결측치가 50% 이상인 컬럼 삭제  
전체 데이터 행의 50% 이상이 결측치인 컬럼은 삭제하세요.  
너무 많은 결측치가 있는 변수는 분석에서 제외하는 것이 일반적입니다.

```python
threshold = len(df) * 0.5
df = df.loc[:, df.isnull().sum() < threshold]
```

```python
df.info()
```

### 데이터 프레임 필터링

전체 데이터프레임 중 앞쪽 18개 열만 선택하여 사용합니다.
iloc을 사용하여 열의 위치 인덱스 기준(0-based)으로 슬라이싱하며,
df.iloc[:, :18]은 0번째부터 17번째 열까지를 포함합니다. 이후 불필요한 열을 제거하고, 분석에 필요한 주요 컬럼들만 남기는 전처리 단계입니다.

```python
df = df.iloc[:,:18] #전체 행은 유지하고, 앞에서부터 18개의 열만 필터링
```

```python
df.head()
```

"최저기온 시각(hhmi) 컬름을 삭제하세요.

```python
df = df.drop(["최저기온 시각(hhmi)"],axis=1)
```

```python
df.head()
```

데이터프레임의 열 이름 중 "평균기온(°C)", "최저기온(°C)", "최고기온(°C)"에 포함된 단위를 제거하고 각각 "평균기온", "최저기온", "최고기온"으로 간결하게 수정합니다.
이는 열 이름을 코드로 다룰 때 가독성을 높이고, 문자열 처리의 불편함을 줄이기 위한 전처리 작업입니다.

```python
df=df.rename(columns={"평균기온(°C)":"평균기온", "최저기온(°C)":"최저기온", "최고기온(°C)":"최고기온"}
```

```python
df.head()
```

### Map함수로 데이터 변환하기

Pandas의 map() 함수는 **Series(열)**에 있는 값들을 함수나 딕셔너리로 변환할 때 사용합니다.
주로 값을 바꾸거나 등급 매기기, 한글 변환, 조건 분류 등에 활용됩니다.

```python
week = {"Monday":"월","Tuesday":"화","Wednesday":"수","Thursday":"목","Friday":"금"}
```

```python
df["요일"] = df["요일"].map(week)
```

```python
df.head()
```

## 데이터 요약 및 통계




### 전체 통행량의 평균, 최솟값, 최댓값, 중앙값 구하기
전체 차량 통행량(승용차)의 평균, 최소값, 최대값, 중앙값을 각각 구하세요.


```python
df["승용차"].mean()
```

```python
df["승용차"].min()
```

```python
df["승용차"].max()
```

```python
df["승용차"].median()
```

### **Groupby** 월별 전체 통행량 평균 구하기
'월' 컬럼을 기준으로 전체 차량 통행량('합계')의 평균을 구하세요.

```python
df.groupby("월")["합계"].mean()
```

'월','혼잡도'로 그룹화를 하고 전체 차량 통행량('합계')의 평균을 구하세요.

```python
df.groupby(["월","혼잡도"])["합계"].mean()
```

```python
df.groupby(['월','혼잡도']).mean(numeric_only=True)
```

### 요일별 차량 종류별 평균 통행량 구하기
요일을 기준으로 승용차, 버스, 택시의 평균을 구하세요.

```python
df.groupby('요일')[['승용차', '버스', '택시']].mean()
```
